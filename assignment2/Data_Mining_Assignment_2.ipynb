```python
import pyfpgrowth

associations = df['Query'].apply(lambda x: x.split()).tolist()

num_records = len(associations)
print("Number of records:", num_records)
print("Example transaction:", associations[0])
```

```python
sigma = 100
min_support = sigma / num_records
print(f"Minimum support (fraction): {min_support:.4f}")

patterns = pyfpgrowth.find_frequent_patterns(associations, sigma)

num_frequent_itemsets = len(patterns)
max_itemset_size = max(len(itemset) for itemset in patterns)

print(f"Number of frequent itemsets: {num_frequent_itemsets}")
print(f"Maximum size of frequent itemsets: {max_itemset_size}")

support = {key: value / num_records for key, value in patterns.items()}

print("Some frequent itemsets with support:")
for k, v in list(support.items())[:10]:
    print(f"{k}: {v:.4f}")
```

```python
single_items = {k: v for k, v in patterns.items() if len(k) == 1}
most_frequent_item = max(single_items, key=single_items.get)
most_frequent_count = single_items[most_frequent_item]

print("Most frequent item:", most_frequent_item[0])
print("Occurs in queries:", most_frequent_count)
```

Task 3 Effect of Support

```python
num_records = len(df)
min_support_values = [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008,0.009, 0.01]

for min_support in min_support_values:
  support_count = min_support * num_records
  patterns = pyfpgrowth.find_frequent_patterns(associations, support_count)
  support = {key: value / len(df) for key, value in patterns.items()}
  print(f"Minimum support: {min_support}, Support count: {support_count:.0f}")
  print(f'Found {len(patterns)} patterns')
  print('Patterns', patterns)
  print('Support', support)
  print()
```

Task 4 rule regulation

```python
min_support = 0.004   # from Task 3.3
support_count = int(min_support * num_records)
confidence = 0.8

patterns = pyfpgrowth.find_frequent_patterns(associations, support_count)

rules = pyfpgrowth.generate_association_rules(patterns, confidence)
print("Number of rules generated:", len(rules))

for antecedent, (consequent, conf) in rules.items():
    reverse_rule = (consequent,)

    if reverse_rule in rules:
        conf_YX = rules[reverse_rule][1]
    else:
        conf_YX = None

    if conf >= 0.8 and (conf_YX is None or conf_YX < 0.8):
        print("High confidence rule:", antecedent, "->", consequent, "Conf:", conf)
        print("Reverse rule:", consequent, "->", antecedent, "Conf:", conf_YX)
        break
```

Task 6 rule interpretation

```python
min_support = 0.001   # very low to generate more rules
support_count = int(min_support * num_records)
confidence = 0.8

patterns = pyfpgrowth.find_frequent_patterns(associations, support_count)
rules = pyfpgrowth.generate_association_rules(patterns, confidence)

print("Number of rules generated:", len(rules))
```

```python
support = {itemset: count / num_records for itemset, count in patterns.items()}
```

```python
rules_with_lift = []

for antecedent, (consequent, conf) in rules.items():
    itemset = tuple(sorted(antecedent + consequent))
    antecedent_key = tuple(sorted(antecedent))
    consequent_key = tuple(sorted(consequent))

    sup_XY = support.get(itemset, 0)
    sup_X = support.get(antecedent_key, 0)
    sup_Y = support.get(consequent_key, 0)

    if sup_X > 0 and sup_Y > 0:   # avoid division by zero
        lift = conf / sup_Y
        rules_with_lift.append(((antecedent, consequent), conf, lift))

```

```python
rules_by_conf = sorted(rules_with_lift, key=lambda x: x[1], reverse=True)

rules_by_lift = sorted(rules_with_lift, key=lambda x: x[2], reverse=True)

print("\nTop 20 rules by confidence:")
for rule, conf, lift in rules_by_conf[:20]:
    print(rule, "Conf:", conf, "Lift:", lift)

print("\nTop 20 rules by lift:")
for rule, conf, lift in rules_by_lift[:20]:
    print(rule, "Conf:", conf, "Lift:", lift)
```
